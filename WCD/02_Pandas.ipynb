## Carregando e analisando o arquivo 'salarios.csv'

```python
# Usando o método read_csv para ler um CSV, considerando a primeira linha como cabeçalh e mudando o nome das colunas
df = pd.read_csv('salarios.csv', header=1, names = ['nome', 'cargo', 'departamento', 'salario_anual'])
```

```python
# Mostrando os primeiros registros
df.head()
```

```python
# Mostrando as estatísticas
df.describe()
```

```python
# Veja abaixo que todos os elementos são convertidos 
print(df.dtypes)
```

```python
# Vamos calcular a média de salários anui
df.mean()
```

```python
# Vamos simplesmente plotar o DataFrame
df.plot()
```

Veja que esse resultado acima não faz muito sentido, pois estamos plotando salario x id. 
Talvez seja melhor utilizar um histograma

```python
# Plotando um histograma dos salários
df.hist(column='salario_anual')
```

Este gráfico mostra no eixo X o salário. 
Aqui já podemos tirar algumas conclusões, como por exemplo, que poucos empregados ganham mais que 100k, ao passo que a grande maioria está na faixa dos 75k a 100k 

```python
# Vamos dar uma olhada nos últimos registros
df
```

Veja que a última linha está com valores nulos

```python
# Vamos dropar qualquer valor nulo ou vazio
df = df.dropna()
```

```python
# Imprimindo o DataFrame
df
```

```python
# Plotando o mesmo histograma de salários com matplotlib, a partir dos dados de valores
valores = df['salario_anual'].values
plt.hist(valores, bins=30, cumulative=False, log=False)
plt.title('Histograma de Salários')
plt.xlabel('Salários')
plt.ylabel('Probabilidade')
```
